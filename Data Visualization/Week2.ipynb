```python
mydata3 = pd.read_csv('ExcelFormattedGISTEMPData2CSV.csv')
mydata3.tail()
```

```python
mydata3.head()
```

```python
nhem = mydata3[['Year','Glob','SHem','NHem']]
nhem.head()
```

```python
plt.plot(nhem['Year'],nhem['NHem'])
plt.plot(nhem['Year'],nhem['SHem'])
```

```python
fig = plt.figure()
ax = plt.subplot(111)
ax.plot(nhem['Year'],nhem['NHem'], label='North Hem temperature')
ax.plot(nhem['Year'],nhem['SHem'], label='South Hem temperature')
#ax.plot(nhem['Year'],nhem['Glob'], label='Global temperature')
plt.title('Temperature of South Hem, North Hem and global from 1880-2019')
ax.legend()
plt.show()
```

```python
# find out which year the South Hem is cooler than North Hem
south_cooler = nhem[nhem['SHem'] < nhem['NHem']]
south_cooler.size
```
